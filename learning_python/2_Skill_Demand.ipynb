What are the most demanded skills for the top 3 most popular data roles?

```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#clean data to daetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_skills = df_US.explode('job_skills').copy()

df_skills[['job_title_short', 'job_skills']]
```

```python
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

df_skills_count = df_skills_count.reset_index(name='skill_count').copy()
```

```python
#sorting data 
df_skills_count.sort_values('skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles  = sorted(job_titles[:3])

job_titles
```

```python
#creating viz 
fig, ax = plt.subplots(len(job_titles), 1)

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('Top 5 Skills for Trending Job Titles', fontsize=16, fontweight='bold')
fig.tight_layout(h_pad=0.5)
plt.show()
```

```python
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='job_total').copy()

df_job_title_count
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, how= 'left', on='job_title_short').copy()

df_skills_perc['skill_percent'] = 100 * df_skills_perc['skill_count'] / df_skills_perc['job_total'] 

df_skills_perc['skill_percent'] = df_skills_perc['skill_percent'].round(2)

df_skills_perc
```

```python
#creating viz 
fig, ax = plt.subplots(len(job_titles), 1)

sns.set_theme(style='ticks')


for i, job_title in enumerate(job_titles):
    df_plot_2 = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot_2, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_percent', palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].set_xlim(0,100)
    ax[i].legend().set_visible(False)

    for n, v in enumerate(df_plot_2['skill_percent']):
        ax[i].text(v + 1, n, f"{v:.0f}%", va='center')

    if i != len(job_titles) -1:
        ax[i].set_xticks([])

fig.suptitle('Likelihood of Skills Requested in US Job Posting', fontsize=16, fontweight='bold')
fig.tight_layout(h_pad=0.5)
plt.show()
```
